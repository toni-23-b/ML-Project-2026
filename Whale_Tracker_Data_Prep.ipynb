```python
import pandas as pd

# 1. Paste your Google Drive file path here!
file_path = '/content/drive/MyDrive/Binance_ETHUSDT_1h (1).csv'
print("Loading the dataset...")
df_all_crypto = pd.read_csv(file_path)

# 2. THE FIX: Strip any accidental hidden spaces from the column names
df_all_crypto.columns = df_all_crypto.columns.str.strip()

# 3. Filter using the 'Symbol' column (Searching for ETHUSDT)
print("Filtering for Ethereum only...")
df_eth = df_all_crypto[df_all_crypto['Symbol'] == 'ETHUSDT'].copy()

# 4. Clean the Date column
print("Cleaning the timestamps...")
df_eth['Date'] = df_eth['Date'].astype(str).str.replace(r'\.\d{3}', '', regex=True)
df_eth['Date'] = pd.to_datetime(df_eth['Date'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# 5. Show the results!
print("\n--- Success! Here is your clean Ethereum Data ---")
display(df_eth.head())
```

```python
# 1. Paste your new Whale Data Google Drive path here!
whale_file_path = '/content/drive/MyDrive/ML_crypto/bigquery_whale_data.csv'

print("Loading the Whale data from BigQuery...")
df_whales = pd.read_csv(whale_file_path)

# 2. Convert the BigQuery timestamp into official Time data
# (We must do this so it matches the Kaggle Date column!)
df_whales['hour_timestamp'] = pd.to_datetime(df_whales['hour_timestamp'], errors='coerce')

# 3. Show the results
print("\n--- Success! Here is your BigQuery Whale Data ---")
display(df_whales.head())
```
