```python
import pandas as pd

# 1. Paste your Google Drive file paths here!
kaggle_path = '/content/drive/MyDrive/ML_crypto/Binance_ETHUSDT_1h (1).csv'
whale_path = '/content/drive/MyDrive/ML_crypto/bigquery_whale_data.csv'

# 2. Load and prep Kaggle Price Data
print("Loading Price Data...")
df_prices = pd.read_csv(kaggle_path)
df_prices.columns = df_prices.columns.str.strip()
df_eth = df_prices[df_prices['Symbol'] == 'ETHUSDT'].copy()
df_eth['Date'] = df_eth['Date'].astype(str).str.replace(r'\.\d{3}', '', regex=True)
df_eth['Date'] = pd.to_datetime(df_eth['Date'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# 3. Load and prep BigQuery Whale Data
print("Loading Whale Data...")
df_whales = pd.read_csv(whale_path)
df_whales['hour_timestamp'] = pd.to_datetime(df_whales['hour_timestamp'], errors='coerce')

# THE FIX: Remove the UTC timezone from BigQuery so it matches Kaggle
df_whales['hour_timestamp'] = df_whales['hour_timestamp'].dt.tz_localize(None)

# 4. THE MERGE (Stitching them together)
print("Merging datasets together...")
final_df = pd.merge(df_eth, df_whales, left_on='Date', right_on='hour_timestamp', how='inner')

# 5. Clean up duplicate columns and sort by time
final_df = final_df.drop(columns=['hour_timestamp']) # Drop the extra time column
final_df = final_df.sort_values('Date').reset_index(drop=True)

print("\n--- Ultimate Dataset Ready! ---")
print(f"Total Rows: {len(final_df)}")
display(final_df.head())
```

```python
# 1. Choose a name for your final master file
file_name = 'whale_dataset_merged.csv'

# 2. Tell Pandas to convert our 'final_df' table into a CSV file
# (index=False prevents it from adding a useless column of row numbers)
final_df.to_csv(file_name, index=False)
print("File converted to CSV!")

# ---------------------------------------------------------
# OPTION A: Save it directly to your Google Drive folder
# (Replace with your actual Drive folder path if it's different)
# ---------------------------------------------------------
drive_path = f'/content/drive/MyDrive/ML_crypto/{file_name}'
final_df.to_csv(drive_path, index=False)
print(f"Success! File saved to Google Drive at: {drive_path}")

```
